```python
clrs = []
for clr in df_sales.color.str.split():
    clrs.extend(clr)
    
text = ' '.join(clrs)
```

```python
word_cloud = WordCloud(collocations=False, background_color='white').generate(text)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
ctr = Counter(text.split())
top_nine_colors = ctr.most_common(9)
print(top_nine_colors)
```

```python
clr_to_check = [x for (x,y) in top_nine_colors]
```

```python
#If there are if statments, read the if statements first

df_sales['base_colors'] = df_sales['color'].str.split().apply(lambda x:[i for i in x if i in clr_to_check]) #midnight black, removes midnight and gives black and black is a list
df_sales['base_colors'] = df_sales['base_colors'].apply(lambda x: ['others'] if x==[] else x) #replaces emply lists with others
df_sales['base_colors'] =  df_sales['base_colors'].apply(lambda x:' '.join(x)) 
```

```python
df = pd.DataFrame(np.round(df_sales['base_colors'].value_counts(normalize=True)*100, 2)).reset_index()
df.columns=['Base_colors','count_perc']
df.head(10)
```
